# Advanced Business Intelligence Analysis

This notebook extends KPI analysis into deeper business insights using Python-based analytical techniques to simulate real-world business intelligence workflows.

## Objective

The objective of this notebook is to go beyond KPI tracking and analyze deeper business patterns across customers, products, revenue, and operations.

We aim to identify:
- High-value customers
- Revenue drivers
- Operational bottlenecks
- Product performance patterns
- Business growth opportunities

```python
import pandas as pd
import numpy as np
```

```python
dim_product = pd.read_csv("../data/processed/dim_product.csv")
dim_customer = pd.read_csv("../data/processed/dim_customer.csv")
dim_hub = pd.read_csv("../data/processed/dim_hub.csv")

fact_orders = pd.read_csv("../data/processed/fact_orders.csv")
fact_delivery = pd.read_csv("../data/processed/fact_delivery.csv")
fact_order_product = pd.read_csv("../data/processed/fact_order_product.csv")
```

## Q1: Who are the highest revenue generating customers?

```python
top_customers = (
    fact_orders.groupby("user_id")["order_revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_customers
```

### Insight

A small group of customers contributes disproportionately to total revenue, indicating strong revenue concentration among high-value users.

These users should be targeted for loyalty programs and retention strategies.

## Q2: Do a small number of customers contribute most of the revenue?

```python
customer_revenue = (
    fact_orders.groupby("user_id")["order_revenue"]
    .sum()
    .sort_values(ascending=False)
)

top_20_percent = int(len(customer_revenue) * 0.2)

top_20_revenue = customer_revenue.head(top_20_percent).sum()
total_revenue = customer_revenue.sum()

top_20_revenue / total_revenue * 100
```

### Insight

The top 20% of customers contribute a significant portion of total revenue, confirming a strong Pareto distribution pattern in customer value.

## Q3: Which products are ordered most frequently?

```python
top_products = (
    fact_order_product.groupby("product_id")["order_id"]
    .count()
    .sort_values(ascending=False)
    .head(10)
)

top_products
```

## Q4: Which product categories generate the most revenue?

```python
category_revenue = (
    fact_order_product
    .merge(dim_product, on="product_id", how="left")
    .groupby("department_id")["order_id"]
    .count()
    .sort_values(ascending=False)
)

category_revenue
```

## Q5: What is the average number of products per order?

```python
basket_size = (
    fact_order_product.groupby("order_id")["product_id"]
    .count()
)

basket_size.mean()
```

### Insight

Customers typically purchase multiple items per order, indicating strong basket-building behavior which is critical for cross-selling strategies.

## Q6: Which time of day generates the highest revenue?

```python
revenue_by_hour = (
    fact_orders.groupby("order_hour_of_day")["order_revenue"]
    .sum()
    .sort_values(ascending=False)
)

revenue_by_hour
```
